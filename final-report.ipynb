# **Comparing life expectancy between low hour workers and high hour workers**

```r
# these libraries will be used, so we load them now
library(digest)
library(infer)
library(repr)
library(taxyvr)
library(tidyverse)
library(dplyr)
library(broom)
library(gt)
library(ggplot2)
```

### **Introduction** 

Historically, it's known that long work hours cause stress, and stress decreases health by introducing symptoms such as depression, anxiety, and poor sleep. (Koenig, 2011; Lockwood, 2003). As such, we want to study if annual hours worked changes life expectancy. Since we will need to split our data into groups, our location parameter is the difference in average life expectancy and our scale parameter is the standard error of the difference. The reproducable dataset can be read from the web and includes information about 44 cities across all continents except Antarctica.

### **Methods and Results** 

This dataset has 12 columns, so we will only need the average life expectancy and the annual average hours worked. We intend to divide the dataset into two groups: low hour workers and high hour workers using the median to divide the group evenly.

Let's first read and clean up the data.

```r
# reading data from web
link <- "https://raw.githubusercontent.com/zhong-test/data-for-class/master/healthy_lifestyle_city_2021.csv"
data <- read.csv(link)
slice(data, 1)
cat("Table 1.0: Shows the original columns of the data.")
```

Our study only requires the life expectancy and hours worked columns so we filter out uneeded columns.

```r
# tidying up the data
tidy <- data |> select(City, `Life.expectancy.years...Country.`,`Annual.avg..hours.worked`) |>
        reframe(Life = `Life.expectancy.years...Country.`, Work = as.numeric(`Annual.avg..hours.worked`)) |> # renamed columns 
        filter(!is.na(Work)) 
slice(tidy, 1)
cat("Table 1.1: Shows the tidied data frame.")
```

Now we'll separate the cities into 2 groups, cities with low work hours and high work hours. This essentially divides the population extrapolated by cities into low hour workers and high hour workers.

```r
# separating the cities into 2 groups, evenly divided by the median
median <- median(tidy$Work)
group <- tidy |> mutate(hi_work = Work > median)
slice(group, 1, 2)
cat("Table 1.3: Shows the sample being split into 2 groups.")
```

We'll now compare the estimated life expectancy between the two groups.

```r
# estimates for average life expectancy between 2 groups and their difference
hi_worker_life <- group |> filter(hi_work == TRUE) |> summarise(mean = mean(Life)) |> pull(mean)
lo_worker_life <- group |> filter(hi_work == FALSE) |> summarise(mean = mean(Life)) |> pull(mean)
sample_mean_diff <- lo_worker_life - hi_worker_life
data.frame(lo_workers_mean = lo_worker_life, hi_workers_mean = hi_worker_life, diff = sample_mean_diff)
cat("Table 1.4: Displays the computed difference in mean lifespan from the sample.")
```

We see an immediate difference in the average lifespans between low hour workers and high hour workers. It is possible that the difference may have arised from the randomness of the sample, so we need a hypothesis test to whether the difference is statistically significant. We will now test the hypothesis at 5% level where $H_0:\mu_1=\mu_2$ and $H_A:\mu_1>\mu_2$ where $\mu_1$ is the average lifespan of low hour workers and $\mu_2$ is the average lifespan of high hour workers. If the p value of the hypothesis test is below 0.05, then we have evidence to reject the idea that the lifespans between the 2 groups are equal. In addition, we will find the 95% confidence interval to see plausible values for the average difference given our sample.

Before we start the hypothesis test, we need to know if the distribution of lifespans is roughly normal and unimodal. We plan to use the t distribution as we don't have the population standard deviation. We will plot the original data to know if we should continue with a t test.

```r
# plotting the original lifespans to determine if a t test is suitable
options(repr.plot.width = 6, repr.plot.height = 3)
group |> ggplot(aes(x=Life)) + geom_histogram(binwidth = 5) + labs(x = "Life Expectancy (Years)", y = "Frequency",  title = "Distribution of Lifespans")
cat("Figure 1.5: Distribution for all lifespans in our sample. It is unimodal and roughly normal, so a t test is suitable.")
```

From the above plot, the lifespans in our sample is roughly normal and is unimodal, so a t test is suitable. We now use the 2-sample, 1-sided independent t distribution with df $\approx$ 15 as the groups are independent from each other and 15 is 1 less than the size of the smallest group. We use the t.test package to solve for the p value and to get a confidence interval as it reduces the chance of manual errors.

```r
# finding the p value of t sample t test using t.test
t.test(x = group %>% filter(hi_work == FALSE) |> select(Life),
       y = group %>% filter(hi_work == TRUE)  |> select(Life),
       alternative='greater') |> tidy()
cat("Table 1.6: The results of a 2 sample independent t test, 1 sided. The p value is shown.")
```

In our hypothesis test, our p value is 0.0061, meaning there is evidence against the null hypothesis. Reporting the p value alone doesn't give the whole picture, so we will compute the confidence interval of the difference in means between the groups. The autogenerated confidence interval assumes that the data is normal, so it uses asympotics. The upperbound is infinity, but that isn't precise, so we'll create a bootstrapped confidence interval using 1000 replicates.

```r
set.seed(2023) # allows for future replication

# finding confidence interval using bootstrapping
diff_means_bootstrap <- group |> specify(response = Life, explanatory = hi_work) |>  
generate(reps = 1000, type = "bootstrap") |> calculate(stat = "diff in means", order = c("FALSE", "TRUE"))

diff_means_ci <- diff_means_bootstrap |> get_confidence_interval(level = 0.95, type = "percentile")
diff_means_ci
cat("Table 1.7: The bootstrapped confidence interval for differences in lifespan between low hour workers and high hour workers is shown. \n")
cat("Our confidence interval excludes 0, so it disagrees with the null hypothesis.")
```

Our bootstrapped confidence interval aims to estimate possible difference in lifespans between the 2 worker groups given our sample. To see how much the confidence interval disagrees with the null hypothesis, we plot the bootstrap distribution highlighting our 95% confidence interval.

```r
# shading our confidence interval on our bootstrap distribution
diff_means_bootstrap |> visualize() + shade_ci(endpoints = diff_means_ci) + geom_vline(xintercept = sample_mean_diff, linetype = "dashed") +
labs(x = "Difference in Means of Life Expectancy (years)", title = "Bootstrap Distribution of difference in Mean Lifespan",
     caption = "Figure 1.8: Illustration for Boostrapped Confidence Interval")
```

The 95% bootstrapped confidence interval says that for all confidence intervals possibly created, about 95% contain the true average difference in lifespan between our groups. Because our confidence interval doesn't contain 0, we are 95% confident that the true mean difference of lifespans between low and high hour workers is between 1.005453 and 4.49026. In addition, we report a p value less than 0.0061<0.05, so we reject the null hypothesis that the average lifespan of low hour workers isn't higher than high hour workers.

Below we apply bootstrapping to set up the hypothesis test and find out the p-value:

```r
set.seed(2023) # allows for future replication

# hypothesis test assuming the bootstrap distribtion. We should only use the null distribtion for a hypothesis test
life_null_dist <- group |> specify(formula = Life ~ hi_work) |> hypothesize(null = "independence") |> generate(reps = 1000, type = "bootstrap") |>
calculate(stat = "diff in means", order = c("FALSE", "TRUE"))

life_null_dist |> get_p_value(obs_stat = 0, direction = "left") # wrong one: life_null_dist |> get_p_value(obs_stat = sample_mean_diff, direction = "right")
```
